```python
import pandas as pd

# Assuming your CSV file uses ';' as the delimiter
data = pd.read_csv('hp.csv', delimiter=';')  # Specify the correct delimiter

df = pd.DataFrame(data)

# Seleksi fitur yang diinginkan
selected_features = df[['Name', 'Brand', 'Price']]

# Menampilkan hasil seleksi fitur
selected_features
```

```python
selected_features['Price'] = pd.to_numeric(selected_features['Price'].str.replace('.', '', regex=False))
```

```python
average_price = selected_features['Price'].mean()
print("Rata-rata Harga Ponsel:", average_price)
```
